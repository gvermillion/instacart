## Import Stuff

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
import warnings
import sqlite3
```

## Environmental Stuff

```python
warnings.filterwarnings('ignore')
sns.set(font_scale=1.5)
%matplotlib inline

instacart_db = '../instacart.db' # Location of DB
conn = sqlite3.connect(instacart_db) # SQL connection
```

## Queries

```python
select_table_names = "SELECT name FROM sqlite_master WHERE type='table'"
select_all_orders = 'SELECT * FROM orders' 
select_all_order_products = 'SELECT * FROM order_products'
select_all_customers = 'SELECT * FROM users_sampled'
select_all_departments = 'SELECT * FROM departments'
select_all_aisles = 'SELECT * FROM aisles'
select_all_products = 'SELECT * FROM products'
```

## To Pandas

```python
res = conn.execute(select_table_names)
tables = [name[0] for name in res]
print(tables)
```

```python
aisles_df = pd.read_sql_query(select_all_aisles,conn)
aisles_df.drop(['index'],axis=1,inplace=True)
aisles_df.head()
```

```python
products_df = pd.read_sql_query(select_all_products,conn)
products_df.drop(['index'],axis=1,inplace=True)
products_df.head()
```

```python
departments_df = pd.read_sql_query(select_all_departments,conn)
departments_df.drop(['index'],axis=1,inplace=True)
departments_df.head()
```

```python
orders_df = pd.read_sql_query(select_all_orders,conn)
orders_df.head()
```

```python
order_products_df = pd.read_sql_query(select_all_order_products,conn)
order_products_df.head()
```

```python
customers_df = pd.read_sql_query(select_all_customers,conn)
customers_df.head()
```

## Functions

### Named

```python
def make_figure(figsize=(12,9)):
    fig = plt.figure(figsize=figsize)
    ax = plt.gca()
    return fig, ax
```

# Assumptions

1. Tables were dumped `today`, i.e. all data is current.

# Tidy Data

In this section, relevant tables are combined to follow tidy data standards.

## Individual Product Orders

```python
# Add user_id column
order_products_df = pd.merge(order_products_df,orders_df,on='order_id')
order_products_df.drop(['order_number', 'order_dow', 'order_hour_of_day',
       'days_since_prior_order'],axis=1,inplace=True)
# Add product name, aisle, department
order_products_df = pd.merge(order_products_df,products_df,on='product_id')
# Add aisle name
order_products_df = pd.merge(order_products_df,aisles_df,on='aisle_id')
# Add department name
order_products_df = pd.merge(order_products_df,departments_df,on='department_id')
order_products_df.head()
```

```python
total_items_ordered = order_products_df.groupby('reordered')['product_id'].count().sum()
```

## Customers

Average and variance for day/hour of order and days since previous order.

```python
customers_df = pd.read_sql_query(select_all_customers,conn)
customers_df.head()
```

```python
tmp = (orders_df.groupby('user_id')['order_dow','order_hour_of_day','days_since_prior_order']
       .agg({'avg':'mean','var':'var'}))
customers_df = pd.merge(customers_df,tmp,on='user_id')
customers_df.columns = ('user_id avg_dow avg_hour avg_days_since var_dow var_hour var_days_since'
                        .split(' '))
```

Total Number Orders

```python
tmp = pd.DataFrame(orders_df.groupby('user_id')['order_number']
                   .agg({'total_orders':'max'})).reset_index()
customers_df = pd.merge(customers_df,tmp,on='user_id')
```

Average and variance of cart size.

```python
tmp = (order_products_df.groupby(['user_id','order_id'])['add_to_cart_order'].max()
       .groupby(['user_id']).agg({'avg_cart_size':'mean','var_cart_size':'var'}))
customers_df = pd.merge(customers_df,tmp,on='user_id')
```

Ratio of reordered products.

```python
a = order_products_df.groupby('user_id')['reordered'].agg({'total_reordered':'sum'})
b = order_products_df.groupby('user_id')['product_id'].agg({'total_items':'count'})
tmp = (pd.DataFrame(a.total_reordered.div(b.total_items),columns=['ratio_reordered'])
       .reset_index())
customers_df = pd.merge(customers_df,tmp,on='user_id')
```

Average and variance of consistency rate.

```python
a = order_products_df.groupby(['user_id','product_id'])['order_id'].agg({'purchased':'count'})
b = order_products_df.groupby(['user_id'])['order_id'].agg({'total_orders':'count'})
c = a.purchased.div(b.total_orders)
tmp = (c.groupby('user_id').agg({'avg_consistency_rate':'mean','var_consistency_rate':'var'})
       .reset_index())
customers_df = pd.merge(customers_df,tmp,on='user_id')
```

Purchase Rate and Lifetime

```python
a = (orders_df.groupby('user_id')['days_since_prior_order']
     .agg({'avg_days_since':'mean','var_days_since':'var'}))
b = (orders_df.groupby('user_id')['days_since_prior_order'].agg({'lifetime':'sum'}))
tmp = (a[['avg_days_since','var_days_since']]
       .div(b.lifetime,axis=0)
       .rename(columns={'avg_days_since':'avg_purchase_rate',
                        'var_days_since':'var_purchase_rate'})
       .reset_index())
customers_df = pd.merge(customers_df,tmp,on='user_id')
customers_df = pd.merge(customers_df,b,on='user_id')
```

## Traffic Analysis

Determine which day(s) and hour(s) of day experience the most sales with respect to new/existing sale.

First, add `new` boolean categorical based on `1: first_number == 1` and `2: Else`.

```python
orders_df['new'] = 0
first_order_criteria = orders_df['order_number'] == 1
orders_df.loc[first_order_criteria,'new'] = 1
orders_df.columns
```

Group by day and hour.

```python
sales_by_day_hour = (pd.DataFrame(orders_df.groupby(['order_dow','order_hour_of_day'])
                                  .size())
                     .reset_index())
sales_by_day_hour.columns = ['Day','hour','count']
tmp = (pd.DataFrame(orders_df.groupby(['order_dow','order_hour_of_day'])['new']
                    .sum())
       .reset_index())
tmp.columns = ['Day','hour','new']
sales_by_day_hour = pd.merge(sales_by_day_hour,tmp,on=['Day','hour'])
sales_by_day_hour['existing'] = sales_by_day_hour['count'] - sales_by_day_hour['new']
```

```python

fig,ax = make_figure()
g = sns.lineplot(x='hour',y='count',hue='Day',data=sales_by_day_hour,ax=ax,legend='full')
ax.set(xlabel='Hour',ylabel='No. Orders');
```

```python
fig,ax = make_figure()
sns.barplot(x='Day',y='count',data=sales_by_day_hour,ax=ax)

ax.set(xlabel='Day',ylabel='Avg. No. Orders');
```

The busiest days of the week are **Day 0 and Day 1**.

For all days, most orders occur between approx. ***09:00-17:00***.

### For Existing Customers

### Conclusions

1. Traffic trends hold regardless of whether customer is new or existing.
1. **Days 0,1** are generally the busiest days.
1. **Days 2-6** generally experience the same amount of traffic.
1. Fairly universally, traffic is heaviest from ***09:00 - 17:00***.

### Top Products

How many entries to include in Top N lists.

```python
top_n = 10
```

**By Department**

```python
def plot_top_n(category=None,data=None,top_n=top_n,ax=ax,rot=45):
    category = category.lower()
    (data.groupby(category)['product_id']
     .agg({'count':'count'})
     .sort_values('count',ascending=False)
     .iloc[:top_n,:]
     .div(total_items_ordered)
     .plot(kind='bar',ax=ax,legend=False))
    ax.set_xticklabels(ax.get_xticklabels(), rotation=rot);
    category = ' '.join(category.split('_'))
    ax.set(xlabel=category.capitalize(),ylabel='Percent of Total No. Items Purchased [x100]');
```

```python
fig,ax = make_figure()
plot_top_n(category='department',data=order_products_df,ax=ax)
```

**By Aisle:**

```python
fig,ax = make_figure()
plot_top_n(category='aisle',data=order_products_df,ax=ax,rot=60)
```

**By Product:**

```python
fig,ax = make_figure()
plot_top_n(category='product_name',data=order_products_df,ax=ax)
```

### Breakdown: New vs Reorder

```python
fig,ax = make_figure()
tmp = (order_products_df.groupby(['department','reordered'])['product_id']
       .agg({'count':'count'})
       .div(total_items_ordered)
       .reset_index())
sns.barplot(x='department',y='count',hue='reordered',data=tmp)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90);
ax.set(xlabel='Department',ylabel='Ratio of Total Items Sold');
leg = ax.legend()
leg.set_title('Reordered')
```

**Observations:**

        - Diary, Eggs
